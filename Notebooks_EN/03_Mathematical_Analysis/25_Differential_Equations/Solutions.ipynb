# **Solution to First-Order Ordinary Differential Equations**

## **1. Solve \( y'(x) = y(x) \)**

This is a separable differential equation. We can rewrite it as:

$
\frac{dy}{dx} = y
$

Separate the variables:

$
\frac{dy}{y} = dx
$

Now, integrate both sides:

$
\int \frac{1}{y} \, dy = \int 1 \, dx
$

$
\ln |y| = x + C
$

Exponentiate both sides to solve for $ y $:

$
y = Ce^x
$

where $ C $ is a constant of integration.

---

## **2. Solve $ y'(x) = \frac{1}{2} y(x) $**

This is also a separable differential equation. We can rewrite it as:

$
\frac{dy}{dx} = \frac{1}{2} y
$

Separate the variables:

$
\frac{dy}{y} = \frac{1}{2} dx
$

Now, integrate both sides:

$
\int \frac{1}{y} \, dy = \int \frac{1}{2} \, dx
$

$
\ln |y| = \frac{x}{2} + C
$

Exponentiate both sides to solve for $ y $:

$
y = Ce^{\frac{x}{2}}
$

where $ C $ is a constant of integration.

---

## **3. Solve $ \frac{dy}{dx} = x y $**

This is a separable differential equation. We can rewrite it as:

$
\frac{dy}{dx} = x y
$

Separate the variables:

$
\frac{dy}{y} = x \, dx
$

Now, integrate both sides:

$
\int \frac{1}{y} \, dy = \int x \, dx
$

$
\ln |y| = \frac{x^2}{2} + C
$

Exponentiate both sides to solve for \( y \):

$
y = Ce^{\frac{x^2}{2}}
$

where $ C $ is a constant of integration.

---

# **Solution to Second-Order Ordinary Differential Equations**

## **1. Solve \( y''(x) + y'(x) = 0 \) with boundary conditions \( y(0) = 2 \) and \( y'(0) = -1 \)**

This is a linear second-order homogeneous ODE. First, solve the characteristic equation:

$
r^2 + r = 0
$

Factor the equation:

$
r(r + 1) = 0
$

Thus, \( r = 0 \) or \( r = -1 \). The general solution is:

$
y(x) = C_1 + C_2 e^{-x}
$

Now, use the boundary conditions to solve for the constants.

- $ y(0) = 2 $ gives $ C_1 + C_2 = 2 $
- $ y'(0) = -1 $ gives $ -C_2 = -1 $, so $ C_2 = 1 $

Substitute $ C_2 = 1 $ into the first equation:

$
C_1 + 1 = 2 \quad \Rightarrow \quad C_1 = 1
$

Thus, the solution is:

$
y(x) = 1 + e^{-x}
$

---

## **2. Solve $ y''(x) - y(x) = 0 $ with boundary conditions $ y(0) = 2 $ and $ y'(0) = 0 $**

This is a linear second-order homogeneous ODE. First, solve the characteristic equation:

$
r^2 - 1 = 0
$

Thus, $ r = 1 $ or $ r = -1 $. The general solution is:

$
y(x) = C_1 e^{x} + C_2 e^{-x}
$

Now, use the boundary conditions to solve for the constants.

- $ y(0) = 2 $ gives $ C_1 + C_2 = 2 $
- $ y'(0) = 0 $ gives $ C_1 - C_2 = 0 $, so $ C_1 = C_2 $

Substitute $ C_1 = C_2 $ into the first equation:

$
2C_1 = 2 \quad \Rightarrow \quad C_1 = 1
$

Thus, the solution is:

$
y(x) = e^{x} + e^{-x} = 2 \cosh(x)
$

---

## **3. Solve $ y''(x) = -\omega^2 y(x) $**

This is a standard second-order homogeneous ODE with the characteristic equation:

$
r^2 + \omega^2 = 0
$

Thus, $ r = \pm i \omega $, and the general solution is:

$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$

# **Check if $ \psi(t, x) = A \cos(\omega t + kx) $ is a solution of the wave equation:**

The wave equation is:

$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0
$

First, calculate the second partial derivatives:

$
\frac{\partial^2 \psi(t, x)}{\partial t^2} = -A \omega^2 \cos(\omega t + kx)
$

$
\frac{\partial^2 \psi(t, x)}{\partial x^2} = -A k^2 \cos(\omega t + kx)
$

Now substitute into the wave equation:

$
-A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0
$

Factor out $ -A \cos(\omega t + kx) $:

$
-A \cos(\omega t + kx) (\omega^2 - v^2 k^2) = 0
$

Since $ A \neq 0 $ and $ \cos(\omega t + kx) \neq 0 $, we have:

$
\omega^2 = v^2 k^2
$

Since $v = \frac{\omega}{k} $, the function $ \psi(t, x) = A \cos(\omega t + kx) $ is indeed a solution to the wave equation.
