# **Derivatives and Limits**

## **1. Compute Derivatives**

### **Linear and Constant Functions**
For $ y(x) = -3x + 3 $:

$
y'(x) = -3
$

For $ y(x) = \pi x + \sin(1) $:

$
y'(x) = \pi
$

For $ y(x) = 4 + \sin(2) $ (constant function):

$
y'(x) = 0
$

---

### **Polynomial Functions**
For $ y(x) = 2x^3 - 3x^2 + 8x - 9 $:

$
y'(x) = 6x^2 - 6x + 8
$

For $ y(x) = 6x^{1/3} $ (power rule):

$
y'(x) = 6 \cdot \frac{1}{3} x^{-2/3} = 2x^{-2/3} = \frac{2}{x^{2/3}}
$

---

### **Trigonometric and Exponential Functions**
For $ y(x) = x $:

$
y'(x) = 1
$

For $ y(x) = \cos(x) + \sin(x) $:

$
y'(x) = -\sin(x) + \cos(x)
$

For $ y(x) = 2\sin(x)\cos(x) $ (using $ 2\sin(x)\cos(x) = \sin(2x) $):

$
y'(x) = \cos(2x) \cdot 2 = 2\cos(2x)
$

For $ y(x) = x\sin(x) $ (product rule):

$
y'(x) = x\cos(x) + \sin(x)
$

For $ y(x) = (x+1)(x+1) = (x+1)^2 $:
$
y'(x) = 2(x+1)
$

For $ y(x) = \frac{x}{x+1} $ (quotient rule):

$
y'(x) = \frac{(x+1) \cdot 1 - x \cdot 1}{(x+1)^2} = \frac{(x+1) - x}{(x+1)^2} = \frac{1}{(x+1)^2}
$

For $ y(x) = (x+1)e^x $ (product rule):

$
y'(x) = (x+1)e^x + e^x = (x+2)e^x
$

For $ y(x) = \sin(x^2) $ (chain rule):

$
y'(x) = \cos(x^2) \cdot 2x = 2x\cos(x^2)
$

For $ y(x) = e^{-2x} $ (chain rule):

$
y'(x) = -2e^{-2x}
$

For $ y(x) = \frac{1}{\sin(x+1)} $ (reciprocal rule):

$
y'(x) = -\frac{\cos(x+1)}{\sin^2(x+1)}
$

For $ y(x) = 2x + 1 $:

$
y'(x) = 2
$

## **2. Proofs**
### **Prove $ \frac{d}{dx} (\ln(\sin(x))) = \cot(x) $**

Using the chain rule:

$
\frac{d}{dx} (\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \cos(x) = \cot(x)
$

Thus,

$
\frac{d}{dx} (\ln(\sin(x))) = \cot(x)
$

---

### **Verify that \( f''(x) = -f(x) \) for \( f(x) = \cos(x) \)**

$
f'(x) = -\sin(x)
$

$
f''(x) = -\cos(x)
$
Since $ f''(x) = -f(x) $, the identity is verified.

## **3. Limits using L'HÃ´pital's Rule**
### **Compute $ \lim\limits_{x \to 0} \frac{\sin(x)}{x} $**

Since both numerator and denominator approach $ 0 $:

$
\lim\limits_{x \to 0} \frac{\sin(x)}{x} = \lim\limits_{x \to 0} \frac{\cos(x)}{1} = 1
$

---

### **Compute $ \lim\limits_{x \to \infty} \frac{\ln(x)}{x} $**

Both numerator and denominator approach \( \infty \):

$
\lim\limits_{x \to \infty} \frac{\ln(x)}{x} = \lim\limits_{x \to \infty} \frac{1/x}{1} = \lim\limits_{x \to \infty} \frac{1}{x} = 0
$

---

### **Compute $ \lim\limits_{x \to \infty} \frac{e^x}{x} $**

Since $ e^x $ grows faster than $ x $:

$
\lim\limits_{x \to \infty} \frac{e^x}{x} = \lim\limits_{x \to \infty} \frac{e^x}{1} = \infty
$

## **4. Particle Motion: Velocity and Acceleration**
Given $ x(t) = 3t^2 - 6t + 1 $, find:

### **Velocity**:

$
V(t) = x'(t) = 6t - 6
$

At $ t = 2 $:

$
V(2) = 6(2) - 6 = 12 - 6 = 6
$

### **Acceleration**:

$
a(t) = V'(t) = 6
$

At $ t = 2 $:
$
a(2) = 6
$

Thus, at $ t = 2 $:

$
V(2) = 6, \quad a(2) = 6
$
