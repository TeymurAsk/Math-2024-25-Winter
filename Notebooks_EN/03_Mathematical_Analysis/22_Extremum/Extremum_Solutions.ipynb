# **Optimization and Extrema Problems**

## **1. Maximizing the Profit Function**

The profit function is given as:

$
P(u) = -2u^2 + 50u - 300
$

To find the number of units that maximize profit, we take the derivative:

$
P'(u) = \frac{d}{du} (-2u^2 + 50u - 300) = -4u + 50
$

Setting $ P'(u) = 0 $:

$
-4u + 50 = 0
$

$
u = \frac{50}{4} = 12.5
$

Since the number of units must be an integer, we check **$ u = 12 $ and $ u = 13 $** to find the maximum profit.

Computing second derivative:

$
P''(u) = -4
$

Since $ P''(u) < 0 $, the function is concave down, confirming that $ u = 12.5 $ gives a **maximum profit**. The optimal number of units is **12 or 13**, whichever is practical.

## **2. Largest Rectangle with 10m of String**

We want to maximize the area of a rectangle given a perimeter of **10 meters**.

Let **$ x $** and **$ y $** be the rectangle's length and width. The perimeter equation is:

$
2x + 2y = 10 \quad \Rightarrow \quad x + y = 5
$

Solve for $ y $:

$
y = 5 - x
$

The area function is:

$
A(x) = x \cdot y = x(5 - x)
$

To maximize, take the derivative:
$
A'(x) = 5 - 2x
$

Set $ A'(x) = 0 $:

$
5 - 2x = 0
$

$
x = 2.5
$

Since the second derivative:

$
A''(x) = -2
$

is negative, the function has a **maximum** at $ x = 2.5 $, so the optimal rectangle is **2.5m Ã— 2.5m** (a square).

## **3. Finding Extremum of $ f(x) = x^2 + 3x - 5 $**

Take the derivative:

$
f'(x) = 2x + 3
$

Set $ f'(x) = 0 $:

$
2x + 3 = 0
$

$
x = -\frac{3}{2}
$

To classify, compute the second derivative:

$
f''(x) = 2 > 0
$

Since $ f''(x) > 0 $, the function has a **minimum** at $ x = -\frac{3}{2} $.

Value at extremum:

$
f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5
$

$
= \frac{9}{4} - \frac{9}{2} - 5 = \frac{9}{4} - \frac{18}{4} - \frac{20}{4} = \frac{-29}{4}
$

Thus, the **minimum point** is:

$
\left(-\frac{3}{2}, -\frac{29}{4}\right)
$

## **4. Finding Extremum of $ f(x) = \frac{x^2 + 2x + 1}{x - 1} $**

Rewrite:

$
f(x) = \frac{(x+1)^2}{x-1}
$
Use the quotient rule:

$
f'(x) = \frac{(2(x+1)(x-1) - (x+1)^2)}{(x-1)^2}
$

Factor:

$
f'(x) = \frac{2x^2 - 2x + 2x + 2 - (x^2 + 2x + 1)}{(x-1)^2}
$

$
= \frac{x^2 + 2x + 1}{(x-1)^2}
$

Setting $ f'(x) = 0 $:

$
x^2 + 2x + 1 = 0
$

Factor:

$
(x+1)(x+1) = 0 \quad \Rightarrow \quad x = -1
$

To classify, find second derivative:

$
f''(x) = \frac{(2x+2)(x-1)^2 - 2(x+1)(x^2+2x+1)}{(x-1)^4}
$

Evaluating at $ x = -1 $ gives $ f''(-1) > 0 $, meaning a **minimum** occurs at $ x = -1 $.

Computing $ f(-1) $:

$
f(-1) = \frac{(-1)^2 + 2(-1) + 1}{-1 - 1} = \frac{1 - 2 + 1}{-2} = \frac{0}{-2} = 0
$
Thus, the **minimum point** is:

$
(-1, 0)
