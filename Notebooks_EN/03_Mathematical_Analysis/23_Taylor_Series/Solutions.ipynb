# **Taylor Series and Visualization**

## **1. Taylor Series for $ f(x) = \cos(x) $ around $ x = 0 $ up to the 4th degree**

The Taylor series for $ f(x) = \cos(x) $ around $ x = 0 $ is given by:

$
f(x) = \cos(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} + O(x^6)
$

Thus, the Taylor series up to the 4th degree is:

$
f(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}
$

This is the approximation of $ f(x) $ near $ x = 0 $.

---

## **2. Taylor Series for $ h(x) = \frac{1}{1 - x} $ around $ x = 0 $ up to the 4th degree**

The Taylor series for $ h(x) = \frac{1}{1 - x} $ around $ x = 0 $ is:

$
h(x) = \frac{1}{1 - x} = 1 + x + x^2 + x^3 + x^4 + O(x^5)
$

Thus, the Taylor series up to the 4th degree is:

$
h(x) \approx 1 + x + x^2 + x^3 + x^4
$

This is the approximation of $ h(x) $ near $ x = 0 $.

---

## **3. Taylor Series for $ g(x) = \sin(x) $ around $ x = \pi $ up to the 4th degree**

To find the Taylor series for $ g(x) = \sin(x) $ around $ x = \pi $, we first need to compute the derivatives of $ \sin(x) $ at $ x = \pi $.

- $ g(x) = \sin(x) $ and $ g(\pi) = 0 $
- $ g'(x) = \cos(x) $, so $ g'(\pi) = -1 $
- $ g''(x) = -\sin(x) $, so $ g''(\pi) = 0 $
- $ g'''(x) = -\cos(x) $, so $ g'''(\pi) = 1 $
- $ g^{(4)}(x) = \sin(x) $, so $ g^{(4)}(\pi) = 0 $

Now, the Taylor series up to the 4th degree is:

$
g(x) = \sin(x) \approx (x - \pi) - \frac{(x - \pi)^3}{6}
$

This is the approximation of $ g(x) $ around $ x = \pi $.

## **4. Tangent Line to $ f(x) = e^{\sin(x)} $ at $ x_0 = \pi $**

We are tasked with finding the tangent line to $ f(x) = e^{\sin(x)} $ at $ x_0 = \pi $. The equation of the tangent line is:

$
y = f'(x_0)(x - x_0) + f(x_0)
$

First, we compute the derivatives of \( f(x) \):

- $ f(x) = e^{\sin(x)} $
- $ f'(x) = e^{\sin(x)} \cdot \cos(x) $

Now evaluate at $ x_0 = \pi $:

- $ f(\pi) = e^{\sin(\pi)} = e^0 = 1 $
- $ f'(\pi) = e^{\sin(\pi)} \cdot \cos(\pi) = 1 \cdot (-1) = -1 $

Thus, the equation of the tangent line at \( x_0 = \pi \) is:

$
y = -1(x - \pi) + 1 = -x + \pi + 1
$
