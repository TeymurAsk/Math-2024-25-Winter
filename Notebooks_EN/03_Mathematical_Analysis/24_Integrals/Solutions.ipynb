# **Integral Calculations and Area Under Curves**

## **1. Compute the following integrals:**

### **a)** $ \int 1 \, dx $

$
\int 1 \, dx = x + C
$

### **b)** $ \int (x^2 + 2) \, dx $

$
\int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C
$

### **c)** $ \int 2 \sin(x) \, dx $

$
\int 2 \sin(x) \, dx = -2 \cos(x) + C
$

### **d)** $ \int 3x \, dx $

$
\int 3x \, dx = \frac{3x^2}{2} + C
$

### **e)** $ \int \frac{1}{x^2} \, dx $

$
\int \frac{1}{x^2} \, dx = -\frac{1}{x} + C
$

### **f)** $\int \left( \frac{1}{3} x^4 - 5 \right) \, dx $

$
\int \left( \frac{1}{3} x^4 - 5 \right) \, dx = \frac{1}{15} x^5 - 5x + C
$

### **g)** $ \int \left( \sin^2(x) + \cos^2(x) \right) \, dx $

Using the Pythagorean identity \( \sin^2(x) + \cos^2(x) = 1 \):

$
\int \left( \sin^2(x) + \cos^2(x) \right) \, dx = \int 1 \, dx = x + C
$

### **h)** $ \int \left( 5 \sin(x) + 3 e^x \right) \, dx $

$
\int \left( 5 \sin(x) + 3 e^x \right) \, dx = -5 \cos(x) + 3 e^x + C
$

### **i)** $ \int x^3 \, dx $

$
\int x^3 \, dx = \frac{x^4}{4} + C
$

### **j)** $ \int 10x \, dx $

$
\int 10x \, dx = 5x^2 + C
$

### **k)** $ \int \cos \left( \frac{5}{2} x + 3 \right) \, dx $

Let \( u = \frac{5}{2} x + 3 \), so \( du = \frac{5}{2} dx \), then:

$
\int \cos \left( \frac{5}{2} x + 3 \right) \, dx = \frac{2}{5} \sin \left( \frac{5}{2} x + 3 \right) + C
$

### **l)** $ \int \cos(\ln(x)) \, x \, dx $

Let $ u = \ln(x) \), so \( du = \frac{1}{x} dx $, then:

$
\int \cos(\ln(x)) \, x \, dx = \sin(\ln(x)) + C
$
### **m)** $ \int x \ln(x) \, dx $

Using integration by parts:

$
\int x \ln(x) \, dx = \frac{x^2}{2} \ln(x) - \frac{x^2}{4} + C
$

### **n)** $ \int x e^x \, dx $

Using integration by parts:

$
\int x e^x \, dx = (x - 1) e^x + C
$

## **2. Calculate integrals over the interval \([0, \pi]\) and visualize them in Geogebra**

### **a)** \( f(x) = 2x + 1 \)

The area under \( f(x) \) from \( x = 0 \) to \( x = \pi \) is:

$
\int_0^\pi (2x + 1) \, dx = \left[ x^2 + x \right]_0^\pi = \pi^2 + \pi
$

### **b)** \( g(x) = x^2 \)

The area under \( g(x) \) from \( x = 0 \) to \( x = \pi \) is:

$
\int_0^\pi x^2 \, dx = \left[ \frac{x^3}{3} \right]_0^\pi = \frac{\pi^3}{3}
$

## **3. Calculate the area of the region bounded by the lines: \( x = 1 \), \( x = 2 \), \( y = 0 \), and \( y = x^2 + 1 \)**

The area is given by:

$
\int_1^2 (x^2 + 1) \, dx = \left[ \frac{x^3}{3} + x \right]_1^2 = \left( \frac{8}{3} + 2 \right) - \left( \frac{1}{3} + 1 \right) = \frac{7}{3}
$

## **4. Calculate the area under the sine curve over the interval \([0, \pi]\)**

The area under the sine curve is given by:

$
P = \int_0^\pi \sin(x) \, dx = \left[ -\cos(x) \right]_0^\pi = -\cos(\pi) + \cos(0) = 2
$

## **5. Calculate the length of the sine curve over the interval \([0, \pi]\)**

The length of the sine curve is given by:

$
L = \int_0^\pi \sqrt{1 + \left( \frac{d}{dx} \sin(x) \right)^2} \, dx = \int_0^\pi \sqrt{1 + \cos^2(x)} \, dx
$

This is a nontrivial integral and may need numerical methods to approximate.

## **6. Find the distance of the moving particle between time \( t = 0 \) and \( t = 2 \)**

The position of the particle is given by \( x(t) = 3t^2 - 6t + 1 \). The velocity is the derivative of the position:

$
v(t) = \frac{d}{dt} (3t^2 - 6t + 1) = 6t - 6
$

Now, the distance traveled is the integral of the speed (the absolute value of velocity):

$
\int_0^2 |6t - 6| \, dt
$

For \( t = 0 \) to \( t = 2 \), \( v(t) \) changes sign at \( t = 1 \), so:

$
\int_0^1 (6 - 6t) \, dt + \int_1^2 (6t - 6) \, dt
$

Calculating both integrals:

$
\int_0^1 (6 - 6t) \, dt = \left[ 6t - 3t^2 \right]_0^1 = 3
$

$
\int_1^2 (6t - 6) \, dt = \left[ 3t^2 - 6t \right]_1^2 = 3
$

Thus, the total distance is:

$
3 + 3 = 6
$
