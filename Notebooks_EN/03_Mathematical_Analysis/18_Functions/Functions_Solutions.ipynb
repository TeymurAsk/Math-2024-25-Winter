# Function Compositions and Inverses

## 1. Checking Function Compositions

We define:

- $ f(x) = e^x $
- $ g(x) = \ln(x) $

Now, we compute the compositions:

### Compute $ f(g(x)) $
$
f(g(x)) = f(\ln x) = e^{\ln x} = x
$

### Compute $ g(f(x)) $
$
g(f(x)) = g(e^x) = \ln(e^x) = x
$

### Observation:
Since both compositions return $ x $, this means that $ f(x) $ and $ g(x) $ are **inverse functions** of each other.

---

## 2. Finding Inverse Functions for Given Sets

### Case 1: Function $ f $ Given as a Set

We have:
$
f = \{(1,7), (2,9), (3,11)\}
$

To find $ f^{-1} $, swap the pairs:

$
f^{-1} = \{(7,1), (9,2), (11,3)\}
$

### Case 2: Function $ f $ with Non-Unique Outputs

We have:
$
f = \{(1,7), (2,7), (3,11)\}
$

Since the value **7 is repeated**, this function is **not one-to-one**, meaning it does **not have an inverse function**.

---

## 3. Finding the Inverse of $ f(x) = x - 1 $

To find $ f^{-1} $, solve for $ x $ in terms of $ y $:

$
y = x - 1
$

Solve for $ x $:

$
x = y + 1
$

Thus, the inverse function is:

$
f^{-1}(x) = x + 1
$
