# Limits and Asymptotes

## 1. Compute $ \lim\limits_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} $

Factor $ x^4 $ in the denominator and $ x^3 $ in the numerator:

$
\frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{x^3(1 + \frac{2}{x})}{x^4(1 - \frac{3}{x})}
$

Rewriting:

$
\frac{1 + \frac{2}{x}}{x(1 - \frac{3}{x})}
$

As $ x \to \infty $, we get:

$
\lim\limits_{x \to \infty} \frac{1 + \frac{2}{x}}{x(1 - \frac{3}{x})} = \lim\limits_{x \to \infty} \frac{1}{x} = 0
$

Thus, the limit is:

$
\lim\limits_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = 0
$

## 2. Compute $ \lim\limits_{x \to 0} \frac{\sin(3x)}{2x + 1} $

Using the small-angle approximation $ \sin(3x) \approx 3x $ for small $ x $, we get:

$
\lim\limits_{x \to 0} \frac{\sin(3x)}{2x + 1} = \lim\limits_{x \to 0} \frac{3x}{2x + 1}
$

Substituting $ x = 0 $:

$
\frac{3(0)}{2(0) + 1} = \frac{0}{1} = 0
$

Thus:

$
\lim\limits_{x \to 0} \frac{\sin(3x)}{2x + 1} = 0
$

## 3. Find the Asymptotes of $ f(x) = \frac{x^2 - 1}{x^2 + 1} $

### **Vertical Asymptotes**:
Vertical asymptotes occur where the denominator is zero. Solve:

$
x^2 + 1 = 0
$

Since $ x^2 + 1 $ is always positive, there are **no vertical asymptotes**.

### **Horizontal Asymptotes**:
For large $ x $, divide by $ x^2 $:

$
\lim\limits_{x \to \infty} \frac{x^2 - 1}{x^2 + 1} = \lim\limits_{x \to \infty} \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}}
$

Since $ \frac{1}{x^2} \to 0 $, we get:

$
\frac{1 - 0}{1 + 0} = 1
$

Thus, the horizontal asymptote is:

$
y = 1
$

## 4. Find the Asymptotes of $ g(x) = \frac{\sin(x)}{x^2 + 1} $

### **Vertical Asymptotes**:
Solve $ x^2 + 1 = 0 $, which has no real solutions.  
Thus, **no vertical asymptotes**.

### **Horizontal Asymptotes**:
For large $ x $, since $ |\sin(x)| \leq 1 $, the function behaves as:

$
- \frac{1}{x^2 + 1} \leq g(x) \leq \frac{1}{x^2 + 1}
$

Since $ \frac{1}{x^2 + 1} \to 0 $ as $ x \to \infty $, by the Squeeze Theorem:

$
\lim\limits_{x \to \infty} g(x) = 0
$

Thus, the horizontal asymptote is:

$
y = 0
$
