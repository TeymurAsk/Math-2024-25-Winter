# **Planes**

### 1. **Planes in Space**

A plane in space can be represented by an equation involving three variables $(x, y, z)$. It is a flat, two-dimensional surface extending infinitely in three-dimensional space.

### 2. **Vector Equation of a Plane**

The **vector equation** of a plane is given as:

$$
\vec{r} \cdot \hat{n} = d
$$

Where:
- $\vec{r}$ is the position vector of any point $(x, y, z)$ on the plane,
- $\hat{n}$ is the unit normal vector to the plane, and
- $d$ is the scalar constant representing the plane's distance from the origin along the normal vector.

Alternatively, the vector equation can be written as:

$$
\vec{r} = \vec{r_0} + t \vec{v_1} + s \vec{v_2}
$$

Where:
- $\vec{r_0}$ is a point on the plane,
- $\vec{v_1}$ and $\vec{v_2}$ are non-parallel vectors lying on the plane, and
- $t$ and $s$ are scalar parameters.

### 3. **General Equation of a Plane**

The general form of the plane equation is:

$$
Ax + By + Cz + D = 0
$$

Where:
- $(A, B, C)$ is the normal vector to the plane,
- $D$ is the constant term.

### 4. **Plane Equation Passing Through Three Points**

Given three points $P_1(x_1, y_1, z_1)$, $P_2(x_2, y_2, z_2)$, and $P_3(x_3, y_3, z_3)$, the plane passing through these points can be found by the cross product of vectors $\overrightarrow{P_1P_2}$ and $\overrightarrow{P_1P_3}$:

1. Find the vectors:
   $$ \vec{v_1} = \overrightarrow{P_1P_2} = (x_2 - x_1, y_2 - y_1, z_2 - z_1) $$
   $$ \vec{v_2} = \overrightarrow{P_1P_3} = (x_3 - x_1, y_3 - y_1, z_3 - z_1) $$

2. The normal vector $\hat{n}$ is given by the cross product:
   $$ \hat{n} = \vec{v_1} \times \vec{v_2} $$

3. The equation of the plane is:
   $$ \hat{n} \cdot (x - x_1, y - y_1, z - z_1) = 0 $$

### 5. **Plane Equation in Intercept Form on the Coordinate Axes**

The intercept form of the plane equation is:

$$
\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1
$$

Where:
- $a$, $b$, and $c$ are the intercepts on the $x$-axis, $y$-axis, and $z$-axis respectively.

### 6. **Special Cases of Plane Positioning**

- **Parallel Planes**: Two planes are parallel if their normal vectors are parallel (i.e., the normal vectors are scalar multiples of each other).
- **Coincident Planes**: Two planes are coincident (the same plane) if they have the same normal vector and the same constant term in the general equation.

### 7. **Angle Between Two Planes**

The angle $\theta$ between two planes is given by the formula:

$$
\cos(\theta) = \frac{\vec{n_1} \cdot \vec{n_2}}{|\vec{n_1}| |\vec{n_2}|}
$$

Where:
- $\vec{n_1}$ and $\vec{n_2}$ are the normal vectors of the two planes.

### 8. **Conditions for Parallelism and Perpendicularity of Two Planes**

- **Parallel Planes**: Two planes are parallel if their normal vectors are parallel:
  $$ \vec{n_1} \parallel \vec{n_2} $$

- **Perpendicular Planes**: Two planes are perpendicular if the dot product of their normal vectors is zero:
  $$ \vec{n_1} \cdot \vec{n_2} = 0 $$

# **Line in Space**

### 9. **Types of Line Equations in Space**

The general form of a line in space can be represented as:

1. **Vector Equation**:

$$
\vec{r} = \vec{r_0} + t \vec{v}
$$

Where:
- $\vec{r}$ is the position vector of any point on the line,
- $\vec{r_0}$ is a fixed point on the line (known as the point of reference),
- $\vec{v}$ is the direction vector, and
- $t$ is a parameter.

2. **Parametric Equations**: The parametric form of a line in space is given by:

$$
x = x_0 + t v_x, \quad y = y_0 + t v_y, \quad z = z_0 + t v_z
$$

Where $(x_0, y_0, z_0)$ is a point on the line, and $(v_x, v_y, v_z)$ are the components of the direction vector.

### 10. **Relative Positions of a Line and a Plane**

- **Intersection**: A line and a plane intersect if the line is not parallel to the plane and the point of intersection satisfies the equation of the plane.
- **Parallel**: A line and a plane are parallel if the direction vector of the line is perpendicular to the normal vector of the plane.
- **Coincident**: A line lies on a plane if every point on the line satisfies the plane equation.

### 11. **Relative Positions of Two Lines**

- **Parallel Lines**: Two lines are parallel if their direction vectors are scalar multiples of each other.
- **Skew Lines**: Two lines are skew if they are not parallel and do not intersect.
- **Intersecting Lines**: Two lines intersect if there exists a solution to their parametric equations.

### 12. **Angle Between a Line and a Plane**

The angle $\theta$ between a line and a plane is given by:

$$
\cos(\theta) = \frac{|\vec{v} \cdot \vec{n}|}{|\vec{v}| |\vec{n}|}
$$

Where:
- $\vec{v}$ is the direction vector of the line,
- $\vec{n}$ is the normal vector of the plane.

$$

$$
\hat{n} = \vec{AB} \times \vec{AC} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 1 & 2 & 3 \\ 2 & 0 & 1 \end{vmatrix} = (-6, 1, -4)
$$

3. The equation of the plane is:

$$
-6(x - 1) + (y - 1) - 4(z - 1) = 0
$$

Given the equations of two planes:

$$
2x + 3y - z = 1
$$

$$

Find the angle between these two planes.

1. Compute the normal vectors:

$$
\vec{n_1} = (2, 3, -1), \quad \vec{n_2} = (1, -4, 2)
$$

2. Compute the cosine of the angle:

$$
\cos(\theta) = \frac{\vec{n_1} \cdot \vec{n_2}}{|\vec{n_1}| |\vec{n_2}|} = \frac{2(1) + 3(-4) + (-1)(2)}{\sqrt{2^2 + 3^2 + (-1)^2} \sqrt{1^2 + (-4)^2 + 2^2}} = \frac{-4}{\sqrt{14} \times \sqrt{21}}
$$

### **1. Equation of a Circle**

Find the equation of a circle with center at point $A(1,2)$ and radius $r=3$.

The general equation of a circle with center at $(h, k)$ and radius $r$ is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

For the given problem, the center is $A(1,2)$ and the radius is $r=3$, so the equation becomes:

$$
(x - 1)^2 + (y - 2)^2 = 3^2
$$

Simplifying the equation:

$$
(x - 1)^2 + (y - 2)^2 = 9
$$

This is the equation of the circle.

```python
import numpy as np
import matplotlib.pyplot as plt

# **1. Equation of a Circle**
# Circle with center A(1, 2) and radius r = 3
center = (1, 2)
radius = 3

theta = np.linspace(0, 2 * np.pi, 300)
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * np.sin(theta)

plt.figure(figsize=(6,6))
plt.plot(x, y, label="Circle: $(x-1)^2 + (y-2)^2 = 9$")
plt.scatter(center[0], center[1], color='red', zorder=5, label="Center (1, 2)")
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Equation of a Circle")
plt.legend()
plt.grid(True)
plt.show()
```

### **2. Equation of a Parabola**

Find the equation of a parabola intersecting the $Ox$ axis at points $x=2$, $x=4$, and passing through point $y(3)=1$.

The general equation of a parabola can be written as:

$$
y = a(x - p)(x - q)
$$

Where $(p, 0)$ and $(q, 0)$ are the points of intersection with the $x$-axis.

Given the intersections at $x=2$ and $x=4$, we have:

$$
y = a(x - 2)(x - 4)
$$

We are also given that the parabola passes through the point $(3, 1)$. Substituting $x=3$ and $y=1$ into the equation:

$$
1 = a(3 - 2)(3 - 4)
$$

Simplifying:

$$
1 = a(1)(-1) \quad \Rightarrow \quad a = -1
$$

Thus, the equation of the parabola is:

$$
y = -(x - 2)(x - 4)
$$

Expanding the equation:

$$
y = -(x^2 - 6x + 8) \quad \Rightarrow \quad y = -x^2 + 6x - 8
$$

This is the equation of the parabola.

```python
import numpy as np
import matplotlib.pyplot as plt
# **2. Equation of a Parabola**
# Parabola: y = -(x-2)(x-4)
x_vals = np.linspace(1, 5, 400)
y_vals = -(x_vals - 2) * (x_vals - 4)

plt.plot(x_vals, y_vals, label=r'$y = -(x-2)(x-4)$')
plt.scatter([3], [1], color='red', zorder=5, label="Point (3, 1)")
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.title("Equation of a Parabola")
plt.legend()
plt.grid(True)
plt.show()
```

### **3. Center of the Ellipse**

Find the center of the ellipse with the equation:

$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$

To find the center, we need to rewrite the equation in standard form. First, we complete the square for both $x$ and $y$ terms.

#### For $x$-terms:
$$
x^2 - 4x \quad \text{(add and subtract 4)} \quad (x - 2)^2 - 4
$$

#### For $y$-terms:
$$
4y^2 - 16y \quad \text{(factor out the 4)} \quad 4(y^2 - 4y) \quad \text{(add and subtract 4 inside the parentheses)} \quad 4((y - 2)^2 - 4)
$$

Now, substitute these into the original equation:

$$
(x - 2)^2 - 4 + 4((y - 2)^2 - 4) + 16 = 0
$$

Simplify:

$$
(x - 2)^2 + 4(y - 2)^2 - 4 - 16 + 16 = 0 \quad \Rightarrow \quad (x - 2)^2 + 4(y - 2)^2 = 4
$$

The equation is now in standard form, and the center of the ellipse is at:

$$
(2, 2)
$$


```python
# **3. Center of the Ellipse**
# Ellipse: (x-2)^2 + 4(y-2)^2 = 4
theta = np.linspace(0, 2 * np.pi, 300)
x_vals = 2 + 2 * np.cos(theta)
y_vals = 2 + np.sin(theta)

plt.plot(x_vals, y_vals, label=r'$(x-2)^2 + 4(y-2)^2 = 4$')
plt.scatter(2, 2, color='red', zorder=5, label="Center (2, 2)")
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Center of the Ellipse")
plt.legend()
plt.grid(True)
plt.show()
```

### **4. Slope of the Tangent Line to a Circle**

Find the slope ($m>0$) of the line $y = mx - 5$ that is tangent to the circle with the equation:

$$
x^2 + y^2 = 1
$$

For the line to be tangent to the circle, the distance from the center of the circle $(0, 0)$ to the line must be equal to the radius of the circle, which is 1.

The general formula for the distance from a point $(x_1, y_1)$ to a line $Ax + By + C = 0$ is:

$$
\text{Distance} = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

The equation of the line is $y = mx - 5$, which can be rewritten as:

$$
mx - y - 5 = 0
$$

Thus, $A = m$, $B = -1$, and $C = -5$. The distance from the center $(0, 0)$ to the line is:

$$
\frac{|m(0) - 1(0) - 5|}{\sqrt{m^2 + (-1)^2}} = \frac{5}{\sqrt{m^2 + 1}}
$$

Setting this equal to the radius 1:

$$
\frac{5}{\sqrt{m^2 + 1}} = 1
$$

Squaring both sides:

$$
\frac{25}{m^2 + 1} = 1 \quad \Rightarrow \quad m^2 + 1 = 25 \quad \Rightarrow \quad m^2 = 24 \quad \Rightarrow \quad m = \sqrt{24} = 2\sqrt{6}
$$

Thus, the slope of the tangent line is:

$$
m = 2\sqrt{6}
$$


```python
# **4. Slope of the Tangent Line to a Circle**
# Circle: x^2 + y^2 = 1, tangent line: y = 2√6 x - 5
theta = np.linspace(0, 2 * np.pi, 300)
x_vals = np.cos(theta)
y_vals = np.sin(theta)

# Tangent line: y = 2√6 x - 5
x_line = np.linspace(-1, 1, 300)
y_line = 2 * np.sqrt(6) * x_line - 5

plt.plot(x_vals, y_vals, label=r'$x^2 + y^2 = 1$ (Circle)')
plt.plot(x_line, y_line, label=r'$y = 2\sqrt{6}x - 5$ (Tangent Line)', linestyle='--')
plt.scatter([0], [1], color='red', label="Point on Circle (0, 1)")
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Tangent Line to the Circle")
plt.legend()
plt.grid(True)
plt.show()
```

### **5. Intersection Points of Hyperbola and Ellipse**

Find the intersection points of the hyperbola:

$$
x^2 - y^2 = 1
$$

and the ellipse's line:

$$
x^2 + 4y^2 = 6
$$

To find the intersection, solve the two equations simultaneously. Start by solving one of the equations for $x^2$ and substituting it into the other equation.

From the hyperbola:

$$
x^2 = y^2 + 1
$$

Substitute this into the ellipse equation:

$$
(y^2 + 1) + 4y^2 = 6 \quad \Rightarrow \quad 5y^2 + 1 = 6 \quad \Rightarrow \quad 5y^2 = 5 \quad \Rightarrow \quad y^2 = 1
$$

Thus, $y = \pm 1$.

Substitute $y = 1$ into the equation $x^2 = y^2 + 1$:

$$
x^2 = 1 + 1 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
$$

So the intersection points are:

$$
(\sqrt{2}, 1), \quad (-\sqrt{2}, 1)
$$

And for $y = -1$, we get the same $x$ values:

$$
(\sqrt{2}, -1), \quad (-\sqrt{2}, -1)
$$

Thus, the intersection points are:

$$
(\sqrt{2}, 1), \quad (-\sqrt{2}, 1), \quad (\sqrt{2}, -1), \quad (-\sqrt{2}, -1)
$$


```python
# **5. Intersection Points of Hyperbola and Ellipse**
# Hyperbola: x^2 - y^2 = 1, Ellipse: x^2 + 4y^2 = 6
x_vals = np.linspace(-3, 3, 400)
y_vals_hyperbola_pos = np.sqrt(x_vals**2 - 1)
y_vals_hyperbola_neg = -np.sqrt(x_vals**2 - 1)

# Ellipse
y_vals_ellipse_pos = np.sqrt((6 - x_vals**2) / 4)
y_vals_ellipse_neg = -np.sqrt((6 - x_vals**2) / 4)

plt.plot(x_vals, y_vals_hyperbola_pos, label=r'$x^2 - y^2 = 1$ (Hyperbola)', color='green')
plt.plot(x_vals, y_vals_hyperbola_neg, color='green')
plt.plot(x_vals, y_vals_ellipse_pos, label=r'$x^2 + 4y^2 = 6$ (Ellipse)', color='blue')
plt.plot(x_vals, y_vals_ellipse_neg, color='blue')

plt.scatter([np.sqrt(2), -np.sqrt(2)], [1, 1], color='red', zorder=5, label="Intersection Points")
plt.scatter([np.sqrt(2), -np.sqrt(2)], [-1, -1], color='red')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Intersection of Hyperbola and Ellipse")
plt.legend()
plt.grid(True)
plt.show()
```

### **6. Distance Between Branches of the Hyperbola**

For the hyperbola:

$$
x^2 - y^2 = 1
$$

The branches of the hyperbola are located on the $x$-axis for large values of $|x|$, and the distance between the branches at any point is determined by the $y$-coordinate. The branches are symmetric with respect to the $x$-axis, so the distance between the branches is the distance between the points on the $x$-axis where the hyperbola intersects.

The distance between the branches at $y = 0$ is the distance between the points where $x^2 = 1$, i.e., at $x = \pm 1$.

Thus, the distance between the branches is:

$$
\text{Distance} = 1 - (-1) = 2
$$


```python
# **6. Distance Between Branches of the Hyperbola**
# Hyperbola: x^2 - y^2 = 1
x_vals = np.linspace(-3, 3, 400)
y_vals_hyperbola_pos = np.sqrt(x_vals**2 - 1)
y_vals_hyperbola_neg = -np.sqrt(x_vals**2 - 1)

plt.plot(x_vals, y_vals_hyperbola_pos, label=r'$x^2 - y^2 = 1$ (Hyperbola)', color='green')
plt.plot(x_vals, y_vals_hyperbola_neg, color='green')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Distance Between Branches of the Hyperbola")
plt.legend()
plt.grid(True)
plt.show()
```
